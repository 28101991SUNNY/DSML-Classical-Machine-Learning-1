# Decision tree learning

- Decision tree learning is a supervised learning (Parametric Learning Algorithm ) approach used in statistics, data mining and machine learning. 
- In this formalism, a classification or regression decision tree is used as a predictive model to draw conclusions about a set of observations.



- Tree models where the target variable can take a discrete set of values are called classification trees; in these tree structures, leaves represent class labels and branches represent conjunctions of features that lead to those class labels. 


- Decision trees where the target variable can take continuous values (typically real numbers) are called regression trees.



- Decision trees are among the most popular machine learning algorithms given their intelligibility and simplicity.

- In decision analysis, a decision tree can be used to visually and explicitly represent decisions and decision making. In data mining, a decision tree describes data (but the resulting classification tree can be an input for decision making).





# Defination : 
    
- #### Decision trees use hyperplanes which run parallel to any one of the axis to cut your coordinate system into hyper cuboids.

- A Decision Tree consists of a series of sequential decisions, or decision nodes, on some data set's features. 


- The resulting flow-like structure is navigated via conditional control statements, or if-then rules, which split each decision node into two or more subnodes.


- Leaf nodes, also known as terminal nodes, represent prediction outputs for the model.



- To train a Decision Tree from data means to figure out the order in which the decisions should be assembled from the root to the leaves. 
- New data may then be passed from the top down until reaching a leaf node, representing a prediction for that data point.

### Entropy : (pruning )

    - to measure the impurity of node : we use entropy. 
    - how much randomness is the node containing .
    
--------------------------------------------------
--------------------------------------------------
    
    Let Y discrete Random Variable : 
    
    Y belongs to { y1 y2 y3  .. yk }
    
    in case of binary Y belongs to {1,0}
--------------------------------------------------    
--------------------------------------------------    
    Entropy 
    
    H(Y) =  - sum ( P[yi] * log(P[yi]))
        yi is the classes . 
    

--------------------------------------------------    
--------------------------------------------------
    
    in binary classification : 
    
    Y <- {0,1}
    
    H(Y) = - [ P[class0]*log(P[class0]) + P[class1]*log(P[class1])]
   
--------------------------------------------------
--------------------------------------------------
    
    P(Y = 1) = p
    P(Y = 0) = 1-p
    
    
    H(Y) = - [ plog(p) + (1-p)log(1-p)]
--------------------------------------------------
--------------------------------------------------

--------------------------------------------------
--------------------------------------------------
    In binary classification 
    what is the entropy when P[class +ve] = 50%
                        and P[class -ve] = 50% 

```python
def entropy(p):
#     if 1-p == 0 or p == 0:
#         return 0
    if p*(1-p) == 0:
        return 0
    return - ((p*np.log2(p) + (1-p)*np.log2(1-p)))
```

```python
def entropy(y):
    p = y.value_counts(normalize=True).values
    return -np.sum(p*np.log2(p))
```

```python
entropy(data["Attrition"]) # parent node entropy 
```

```python
# Calculate Weighted entropy of a child node : 

def weighted_entropy_of_childern(feature , labels):
    categories  = feature.unique()
    weighted_entropy = 0.0
    for cat in categories:
        label_category = labels[feature == cat ]
        entropy_of_category = entropy(label_category)
        weighted_entropy += (len(label_category)/len(labels)) * entropy_of_category
    return(weighted_entropy)
            
    
def IG(label, feature):
    rootnode_entropy = entropy(label)
    weighted_entropy = weighted_entropy_of_childern(feature,label)
    return rootnode_entropy - weighted_entropy
```

# Gini Impurity : 

    # Similar to entropy
    # impurity of randomness 

    GI(y) = 1- sum(P[yi]^2)    sum of all label probabilities

```python
def GI(label):
    p = label.value_counts(normalize=True).values
    return 1-np.sum((p)**2)
```

```python
def entropy(p):
#     if 1-p == 0 or p == 0:
#         return 0
    if p*(1-p) == 0:
        return 0
    return - ((p*np.log2(p) + (1-p)*np.log2(1-p)))


p = np.arange(0,1.01,0.01)
entropy_list = []
for i in p:
    entropy_list.append(entropy(i))
    
sns.lineplot(y = entropy_list,
        x = p, label = "Entropy")
plt.xlabel("probability")
plt.ylabel("Entropy | Gini - impurity")


gini = []
for i in np.arange(0,1.01,0.01):
    gini.append(1- ((i)**2 + ((1-i)**2)))

sns.lineplot(x = p,
        y = gini , label = "gini")
plt.legend()

plt.show()
```

```python
# Calculate Weighted gini of a child node : 

def GI(label):
    p = label.value_counts(normalize=True).values
    return 1-np.sum((p)**2)


def weighted_gini_of_childern(feature , labels):
    categories  = feature.unique()
    weighted_gini = 0.0
    for cat in categories:
        label_category = labels[feature == cat ]
        gini_of_category = GI(label_category)
        weighted_gini += (len(label_category)/len(labels)) * gini_of_category
    return(weighted_gini)
            
    
def IG(label, feature):
    rootnode_gini = GI(label)
    weighted_gini = weighted_gini_of_childern(feature,label)
    return rootnode_gini - weighted_gini
```
